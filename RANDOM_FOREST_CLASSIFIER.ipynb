### DIGITS CLASSIFICATION USING SKLEARN RANDOM FOREST CLASSIFIER

**Digits dataset from sklearn**

```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits.data[0]                     # DATA FOR 0 DIGITS SHOWN BELOW
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])        # HERE WE ARE SHOWING MATRIX AS AN IMAGE
                                         # BY USING MATSHOW FUNCTIN FROM MATPLOTLIB
                                         # WE TAKE 4 DIGITS TO SHOW i.e.= 0,1,2,3
```

```python
dir(digits)                        # WE HAVE "TARGET" COLUMN IN OUR DATASET
```

```python
digits.target[:4]                  # WE ONLY TAKING FIRST 4 DIGITS OUT OF 10 AND THIS IS THE CLASSIFICATION LABEL
```

```python
df=pd.DataFrame(digits.data)       # HERE WE STORED DIGITS DATA INTO DATAFRAME(DF)
df.head(10)
```

```python
df["target"]=digits.target            # HERE WE ADDING TARGET DATA INTO NEW COLUMN NAME AS "TARGET"
df.head(10)
```

```python
x= df.drop("target",axis="columns")        # SORTING X AXIS (DROPPING TARGET COLUMN AS IT REPRESENT TO Y AXIS)
x
```

```python
y= df.target                            # SORTING Y AXIS DATA                
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(x)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)                 # MODEL SCORE IS PRETTY GOOD
```

```python
y_predict= model.predict(x_test)          # ASSIGNINIG VARIABLE TO PREDICT DATA
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)            # FINALLY CALLING CONFUSION MATRIX
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)             # HERE IS THE HEATMAP INDICATING WHERE MODEL IS GETTING CONFUSED TO RECOGNIZE DIGIT
                                      # WHERE X AXIS = "PREDICTED RESULTS" & Y AXIS = "ACTUAL RESULTS"
```
